```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("C:/Users/HP/Downloads/data science class/hotel_bookings 2.csv")

city_hotel = df[df['hotel'] == 'City Hotel']
resort_hotel=df[df['hotel'] == 'Resort Hotel']

city_hotel['is_canceled'].value_counts(normalize=True)

#Average_daily_rate(ADR),lead_time & Monthly_Cancellations

resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize = 30)

plt.plot(resort_hotel.index,resort_hotel['adr'], label ='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label ='City Hotel')

plt.legend(fontsize = 20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:/Users/HP/Downloads/data science class/hotel_bookings 2.csv")

plt.figure(figsize=(15, 7))

# Prepare data
data = (df[df['is_canceled'] == 1]
    .groupby('arrival_date_month')['adr']
    .mean()
    .reset_index()
)

# Correct month order
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

# Plot
sns.barplot(
    x='arrival_date_month',
    y='adr',
    data=data,
    hue='arrival_date_month',
    order=month_order,
    palette='bright',
    legend=False
)

plt.title('Average ADR per Month (Canceled Bookings)', fontsize=20)
plt.xlabel('Month')
plt.ylabel('Average ADR')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cancelled bookings
cancelled_data = df[df['is_canceled'] == 1]

cancelled_df_adr = cancelled_data.groupby(
    'reservation_status_date'
)[['adr']].mean()

cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Not cancelled bookings
not_cancelled_data = df[df['is_canceled'] == 0]

not_cancelled_df_adr = not_cancelled_data.groupby(
    'reservation_status_date'
)[['adr']].mean()

not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')

plt.plot(
    not_cancelled_df_adr['reservation_status_date'],
    not_cancelled_df_adr['adr'],
    label='Not Cancelled'
)

plt.plot(
    cancelled_df_adr['reservation_status_date'],
    cancelled_df_adr['adr'],
    label='Cancelled'
)

plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()
plt.show()
```
